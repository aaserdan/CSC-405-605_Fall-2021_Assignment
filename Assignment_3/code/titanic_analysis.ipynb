We are going to study the survival rate of passengers on titanic and what variables affected survival.

Load the dataset in `titanic.xls`. It contains data on all the passengers that travelled on the Titanic.

```python
import pandas as pd
from IPython.core.display import HTML
from simple_colors import *

HTML(filename='../data/titanic.html')
```

```python
# you would need xlrd - pip install xlrd
t_file = pd.ExcelFile('../data/titanic.xls')
t_df = t_file.parse("titanic", header=None)
t_df
```

### Women and children first?

*** 1. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (25 pts)*** 

```python
# Setting the first row into the columns titles by renaming them
t_df.rename(columns = {0: 'pclass', 1: 'survived', 2: 'name', 3: 'sex', 4: 'age', 5: 'sibsp', 
                            6: 'parch', 7: 'ticket', 8: 'fare', 9: 'cabin', 10: 'embarked', 11: 'boat',
                            12: 'body', 13: 'home.dest'}, inplace = True)
                            
temp_t_df = t_df.T.drop(columns=0)
t_df =temp_t_df.T
t_df
```

```python
# Grouping by people who survived (1) and those who did not (0); assuming 1 is survived and 0 is not
survived_passengers = t_df.groupby(by = ['survived'])
survived_passengers.head()

```

```python
# Gets the number of people who survived and did not survived
titanic_surv_and_nonsur = survived_passengers.survived.count()

survived_passengers_count = titanic_surv_and_nonsur[1]
not_survived_passengers_count = titanic_surv_and_nonsur[0]
survived_passengers_count
```

```python
# Gets the number of male and female passengers who survived
survivors_by_sex = survived_passengers.get_group(1).groupby(by = ['sex'])
female_survivors = survivors_by_sex.get_group('female').name.count()
male_survivors = survivors_by_sex.get_group('male').name.count()

print("Female survivors: ", female_survivors, "\nMale survivors:", male_survivors)
```

```python
# Calculates the proportion of female and male that survived
female_survivors_proportion = (female_survivors/survived_passengers_count) * 100
male_survivors_proportion = (male_survivors/survived_passengers_count) * 100

# Prints out the results
print("Female survivors proportion: ", female_survivors_proportion, "%", "\nMale survivors proportion:", male_survivors_proportion, '%')
```

*** 2. Calculate the same proportion, but by class and sex. (25 pts)*** 

```python
# Gets the survivors per class
t_class1 = t_df.loc[t_df['pclass'] == 1]
t_class2 = t_df.loc[t_df['pclass'] == 2]
t_class3 = t_df.loc[t_df['pclass'] == 3]

survivors_class1 = t_class1[t_class1['survived'] == 1].survived.count()
survivors_class2 = t_class2[t_class2['survived'] == 1].survived.count()
survivors_class3 = t_class3[t_class3['survived'] == 1].survived.count()


print('Survivors in PCLASS 1: ', survivors_class1, '\nSurvivors in PCLASS 2: ', survivors_class2,'\nSurvivors in PCLASS 3: ', survivors_class3)
```

```python
# Filtering through data only for those who survived
class_and_sex_survivors = t_df.loc[t_df['survived'] == 1]
class_and_sex_survivors.groupby(['pclass', 'sex']).count()


```

```python
class1_female_survivors = (class_and_sex_survivors.groupby(['pclass', 'sex']).get_group((1,'female')).count())['pclass']
class1_male_survivors = (class_and_sex_survivors.groupby(['pclass', 'sex']).get_group((1,'male')).count())['pclass']

class2_female_survivors = (class_and_sex_survivors.groupby(['pclass', 'sex']).get_group((2,'female')).count())['pclass']
class2_male_survivors = (class_and_sex_survivors.groupby(['pclass', 'sex']).get_group((2,'male')).count())['pclass']

class3_female_survivors = (class_and_sex_survivors.groupby(['pclass', 'sex']).get_group((3,'female')).count())['pclass']
class3_male_survivors = (class_and_sex_survivors.groupby(['pclass', 'sex']).get_group((3,'male')).count())['pclass']
```

```python
proportion_class1_female_survivors = (class1_female_survivors/survivors_class1) * 100
proportion_class1_male_survivors = (class1_male_survivors/survivors_class1) * 100

proportion_class2_female_survivors = (class2_female_survivors/survivors_class2) * 100
proportion_class2_male_survivors = (class2_male_survivors/survivors_class2) * 100

proportion_class3_female_survivors = (class3_female_survivors/survivors_class3) * 100
proportion_class3_male_survivors = (class3_male_survivors/survivors_class3) * 100


print('Proportion of female survivors in 1st class: ', proportion_class1_female_survivors, '%')
print('Proportion of male survivors in 1st class: ', proportion_class1_male_survivors, '%')

print('\nProportion of female survivors in 2nd class: ', proportion_class2_female_survivors, '%')
print('Proportion of male survivors in 2nd class: ', proportion_class2_male_survivors, '%')

print('\nProportion of female survivors in 3rd class: ', proportion_class3_female_survivors, '%')
print('Proportion of male survivors in 3rd class: ', proportion_class3_male_survivors, '%')
```

*** 3. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, class and sex. (40 pts)***

```python
# Makes a new column 'age' in the categories by comparing categorizing the ages
t_df['age_range'] = ''

for i in range(1, len(t_df) + 1):
    if (t_df.age[i] < 14.0):
        t_df.age_range[i] = 'child'
    elif (t_df.age[i] >= 14.0) and (t_df.age[i] < 21.0):
        t_df.age_range[i] = 'adolescent'
    elif (t_df.age[i] >= 21.0) and (t_df.age[i] < 65):
        t_df.age_range[i] = 'adult'
    elif (t_df.age[i] >= 65.0):
        t_df.age_range[i] = 'senior'
    else:
        t_df.age_range[i] = 'NA'

t_df
```

```python
# Masks for those who survived by class, sex, and age category
survived_by_sex_age_class = t_df.loc[t_df['survived'] == 1]
survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).count()
```

```python
# Gets the number of survivors by class, sex, and age range
class1_female_children_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'female', 'NA')).count())['pclass']
class2_female_children_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'female', 'child')).count())['pclass']
class3_female_children_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'female', 'child')).count())['pclass']

class1_male_children_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'male', 'child')).count())['pclass']
class2_male_children_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'male', 'child')).count())['pclass']
class3_male_children_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'male', 'child')).count())['pclass']


class1_female_adolescent_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'female', 'adolescent')).count())['pclass']
class2_female_adolescent_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'female', 'adolescent')).count())['pclass']
class3_female_adolescent_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'female', 'adolescent')).count())['pclass']

class1_male_adolescent_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'male', 'adolescent')).count())['pclass']
class2_male_adolescent_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'male', 'adolescent')).count())['pclass']
class3_male_adolescent_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'male', 'adolescent')).count())['pclass']

class1_female_adults_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'female', 'adult')).count())['pclass']
class2_female_adults_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'female', 'NA')).count())['pclass']
class3_female_adults_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'female', 'adult')).count())['pclass']

class1_male_adults_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'male', 'adult')).count())['pclass']
class2_male_adults_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'male', 'adult')).count())['pclass']
class3_male_adults_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'male', 'adult')).count())['pclass']

class1_female_seniors_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'female', 'senior')).count())['pclass']
class2_female_seniors_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'female', 'NA')).count())['pclass']
class3_female_seniors_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'female', 'NA')).count())['pclass']

class1_male_seniors_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((1,'male', 'senior')).count())['pclass']
class2_male_seniors_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((2,'male', 'NA')).count())['pclass']
class3_male_seniors_survivors = (survived_by_sex_age_class.groupby(['pclass', 'sex', 'age_range']).get_group((3,'male', 'NA')).count())['pclass']
```
